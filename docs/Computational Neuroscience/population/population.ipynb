```python
import  numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style("dark")
sns.set(rc={'figure.figsize':(10, 8)})

from neuron import LIF, NeuronType
from population import Population
from typing import Optional
```

```python
import math


I_constant = lambda x: 4
I_step = lambda x: 5.5 if int(x / 20) % 2 == 0 else 0
I_sin = lambda x: 3 * (math.sin((x) / 3)) + 3.5
I_linear = lambda x: 0.04 * x + 4
I_x2 = lambda x: pow(x - 50, 2) *0.004
```

#### Population with 2 fully connected neurons with different input current

```python
neuron1 = LIF(I=I_constant)
neuron2 = LIF(I=I_sin)
population = Population(neurons=[neuron1, neuron2], connection_ratio=1.0)
population.execute(100)

dt = 0.03125

plt.subplot(2, 2, 1)
plt.plot(list(map(lambda x: x * dt, range(len(neuron1.u)))), list(map(I_constant, list(map(lambda x: x * dt, range(len(neuron1.u)))))), color='red')
plt.gca().set_xlabel('Time (ms)')
plt.gca().set_ylabel('Input Current neuron 1 (nA)')

plt.subplot(2, 2, 2)
plt.plot(list(map(lambda x: x * dt, range(len(neuron2.u)))), list(map(I_sin, list(map(lambda x: x * dt, range(len(neuron2.u)))))), color='blue')
plt.gca().set_xlabel('Time (ms)')
plt.gca().set_ylabel('Input Current neuron 2 (nA)')

plt.subplot(2, 1, 2)
plt.title("Population of 2 fully connected neurons")
plt.plot(list(map(lambda x: dt * x, range(len(neuron1.u)))), neuron1.u)
plt.plot(list(map(lambda x: dt * x, range(len(neuron2.u)))), neuron2.u)
plt.gca().set_ylabel('Membrance Potential (mV)')
plt.gca().set_xlabel('Time (ms)')
plt.legend(['neuron 1', 'neuron 2'])

plt.show()
```

```python
inhibitory_neurons = [LIF(I=I_constant, neuron_type=NeuronType.Inhibitory) for i in range(2)]
excitatory_neurons = [LIF(I=I_linear) for i in range(8)]
neurons = inhibitory_neurons + excitatory_neurons
population = Population(neurons=neurons, connection_ratio=1)
population.execute(100)

plt.figure(figsize=(8, 4))
plt.title("Raster plot")
neurons_plots = []
for index, neuron in enumerate(neurons):
    color = 'blue' if neuron.neuron_type == NeuronType.Excitatory else 'red'
    neurons_plots.append(plt.scatter(neuron.fires, [index] * len(neuron.fires), color=color, marker='s'))

plt.yticks(range(10))
plt.gca().set_xlabel("iteration")
plt.gca().set_ylabel("Neuron")
plt.legend([neurons_plots[-1], neurons_plots[0]], ['Excitatory', 'Inhibitory'])
plt.show()
```

```python
def plot_activity(pop: Population, delta_t: Optional[int] = 50, color: str = "blue") -> None:
    activity = [0] * int(100 / dt)
    for i in range(delta_t, int(100 / dt)):
        sum = 0
        for neuron in pop.neurons:
            fires = np.array(neuron.fires)
            fires = fires[fires <= i]
            fires = fires[fires >= i - delta_t]
            sum += len(fires)
        
        activity[i] = sum / delta_t / len(pop.neurons)

    plt.plot(activity, color=color)
    plt.gca().set_xlabel('iteration')
    plt.gca().set_ylabel('Population Activity')
```
